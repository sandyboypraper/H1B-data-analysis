## Calculating the number of states

```python
f.loc[:,'WORKSITE'] = f.loc[:,'WORKSITE'].apply(lambda re:re.split(',')[1][1:])

def change_NA(re):
    if (re=='NA'):
        return 'MARIANA ISLANDS'
    return re

f.loc[:,'WORKSITE'] = f.loc[:,'WORKSITE'].apply(lambda re:change_NA(re))

print(len(f['WORKSITE'].unique()))
```

```python
f.head()
```

```python
f.rename(columns = {'EMPLOYER_NAME':'EMPLOYER','FULL_TIME_POSITION':'FULL_T','PREVAILING_WAGE':'PREV_WAGE',
                     'WORKSITE':'STATE','lon':'LON','lat':'LAT'}, inplace = True)

columns_to_keep = ['CASE_STATUS','YEAR','STATE','SOC_NAME','JOB_TITLE','FULL_T','PREV_WAGE','EMPLOYER','LON','LAT']

f = f[columns_to_keep]
f.columns
```

```python
f['LON'] = f['LON'].apply(lambda lon:float("%.2f"%lon))
f['LAT'] = f['LAT'].apply(lambda lon:float("%.2f"%lon))
```

### calculation the petitions distribution by status

```python
status_freq = [0]*7

statuses = ['CERTIFIED-WITHDRAWN', 'WITHDRAWN', 'CERTIFIED', 'DENIED',
       'REJECTED', 'INVALIDATED',
       'PENDING QUALITY AND COMPLIANCE REVIEW - UNASSIGNED']

for i in range(0,7):
    status_freq[i] = f[f.CASE_STATUS==statuses[i]]['CASE_STATUS'].count()
status_freq    
```

```python
from matplotlib.pyplot import pie,axis,show
import matplotlib as mpl

plt.figure(figsize = (4.5,4.5))
plt.title('PETITIONS BY CASE STATUS')
axis('equal');
pie(status_freq[:4], labels=statuses[:4]);
show()
f.head()
```

## calculation petition distribution by year

```python
years = [2011,2012,2013,2014,2015,2016]
year_count = [0]*6
for i in range(0,6):
    year_count[i] = f[f.YEAR==years[i]]['YEAR'].count()
year_count    
```

```python
sn.set_context("notebook",font_scale=1.0)
plt.figure(figsize=(13,3))
plt.title('PETITIONS DISTRIBUTION BY YEAR')
sn.countplot(f['YEAR'])
```

## Analyse Denied Petitions

```python
denied = f[f.CASE_STATUS=='DENIED']
len(denied)
```

```python
del denied['CASE_STATUS']
denied = denied.reset_index()
denied.head(2)
```

```python
denied_year_count = [0]*6
for i in range(0,6):
    denied_year_count[i] = denied[denied.YEAR==years[i]]['YEAR'].count()
denied_year_count    
```

```python
sn.set_context("notebook",font_scale=1.0)
plt.figure(figsize=(13,3))
plt.title('DENIED PETITIONS DISTRIBUTION BY YEAR')
sn.countplot(denied['YEAR'])
```

#### calculating the rate at which the petitions are denied per year

```python
denied_year_rate = [0]*6
for i in range(0,6):
    denied_year_rate[i] = float("%.2f" % ((denied_year_count[i]/year_count[i])*100))

ratio = pd.DataFrame()
ratio['year'] = years
ratio['denied rate %'] = denied_year_rate
ratio = ratio.set_index(['year'])
ratio.T
    
```

```python
ratio = ratio.reset_index()
sn.set_context("notebook",font_scale=1.0)
plt.figure(figsize=(13,3))
plt.title('DENIED PETITIONS RATE BY YEAR')
g = sn.barplot(x = 'year' , y = 'denied rate %',data = ratio)
```

```python
us_states = ['MICHIGAN', 'TEXAS', 'NEW JERSEY', 'COLORADO', 'MISSOURI',
       'FLORIDA', 'CALIFORNIA', 'VIRGINIA', 'PENNSYLVANIA', 'WISCONSIN',
       'ILLINOIS', 'DISTRICT OF COLUMBIA', 'NORTH CAROLINA',
       'RHODE ISLAND', 'OHIO', 'WASHINGTON', 'GEORGIA', 'NEW YORK',
       'INDIANA', 'MARYLAND', 'MASSACHUSETTS', 'NEVADA', 'NEW HAMPSHIRE',
       'LOUISIANA', 'MINNESOTA', 'SOUTH CAROLINA', 'ARIZONA', 'TENNESSEE',
       'HAWAII', 'UTAH', 'DELAWARE', 'MARIANA ISLANDS', 'KANSAS',
       'CONNECTICUT', 'MAINE', 'OKLAHOMA', 'ARKANSAS', 'PUERTO RICO',
       'ALASKA', 'IOWA', 'OREGON', 'KENTUCKY', 'NEBRASKA', 'NEW MEXICO',
       'WEST VIRGINIA', 'SOUTH DAKOTA', 'NORTH DAKOTA', 'MISSISSIPPI',
       'ALABAMA', 'VERMONT', 'IDAHO', 'WYOMING', 'MONTANA']
petitions_by_states = [0]*53
for i in range (0,53):
    petitions_by_states[i] = f[f.STATE==us_states[i]]['STATE'].count()
pet_state = pd.DataFrame()
pet_state['STATE'] = us_states
pet_state['FILED PETITIONS'] = petitions_by_states
print(sum(petitions_by_states))
```

```python
sn.set_context("notebook",font_scale=1.0)
plt.figure(figsize=(13,5))
plt.title('FILED PETITION BY STATE')
v = sn.barplot(x = 'STATE',y ='FILED PETITIONS', data = pet_state)
rotg = v.set_xticklabels(v.get_xticklabels(),rotation = 90)
```

## no of petitions denied by the state

```python
denied_by_state = [0]*53
for i in range (0,53):
    denied_by_state[i] = denied[denied.STATE==us_states[i]]['STATE'].count()
den_state = pd.DataFrame()
den_state['STATE'] = us_states
den_state['DENIED PETITIONS'] = denied_by_state
print(sum(denied_by_state))
```

```python
sn.set_context("notebook",font_scale=1.0)
plt.figure(figsize=(13,5))
plt.title('DENIED PETITIONS BY STATES')
v = sn.barplot(x = 'STATE',y ='DENIED PETITIONS', data = den_state)
rotg = v.set_xticklabels(v.get_xticklabels(),rotation = 90)
```

#### Rate at ehich the petitions are denied by the state

```python
denied_state_rate = [0]*53
for i in range (0,53):
    denied_state_rate[i] = float("%.2f" % ((denied_by_state[i]/petitions_by_states[i])*100))
ratios = pd.DataFrame()
ratios['STATE'] = us_states
ratios['DENIED PETITIONS rate'] = denied_state_rate
print(sum(denied_state_rate))
```

```python
sn.set_context("notebook",font_scale=1.0)
plt.figure(figsize=(13,5))
plt.title('DENIED PETITIONS RATE IN % BY STATES')
v = sn.barplot(x='STATE' , y = 'DENIED PETITIONS rate',data = ratios)
rotg = v.set_xticklabels(v.get_xticklabels(),rotation = 90)
```

```python
pet_state['DENIED PETITIONS'] = denied_by_state
pet_state['DENIED PETITIONS %'] = denied_state_rate
pet_state = pet_state.sort_values(by='DENIED PETITIONS %',ascending = False)
pet_state
```
